```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
current_climate = np.array([[0.8, 0.3], [0.5, 0.6]])
future_climate = np.array([[0.9, 0.2], [0.5, 0.5]])



current_landuse = np.array([[0, 1], [1, 0]])
future_landuse = np.array([[0, 1], [0, 1]])

current_climate_current_landuse = current_climate * current_landuse
future_climate_current_landuse = future_climate *current_landuse
climate_only = future_climate_current_landuse - current_climate_current_landuse

future_climate_future_landuse = future_climate * future_landuse
climate_landuse=future_climate_future_landuse - current_climate_current_landuse

land_use_only =climate_landuse- climate_only
```

```python
fig, axes = plt.subplots(1, 3, figsize=(9, 2.5))


sns.heatmap(future_climate_current_landuse, ax=axes[0], cmap='Reds', annot=True, cbar=False, fmt='.2f', annot_kws={"size": 12}, square=True)
axes[0].set_title('Future Climate & Current Land Use ', fontsize=12)

sns.heatmap(current_climate_current_landuse, ax=axes[1], cmap='Reds', annot=True, cbar=False, fmt='.2f', annot_kws={"size": 12}, square=True)
axes[1].set_title('Current Climate & Current Land Use', fontsize=12)

# Plot SSP4-RCP6.0 heatmaps
sns.heatmap(climate_only, ax=axes[2], cmap='Reds', annot=True, cbar=False, fmt='.2f', annot_kws={"size": 12}, square=True)
axes[2].set_title('Climate Change only', fontsize=12)


# Add equal signs between the plots
fig.text(0.33, 0.5, '-', fontsize=16, ha='center', va='center')
fig.text(0.65, 0.5, '=', fontsize=16, ha='center', va='center')

# Adjust layout and remove ticks
for ax in axes.flat:
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout(rect=[0, 0, 1, 0.92])
plt.show()
```
