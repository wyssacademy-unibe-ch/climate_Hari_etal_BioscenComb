```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
current_climate = np.array([[0.8, 0.3], [0.5, 0.6]])
future_climate = np.array([[0.9, 0.2], [0.5, 0.5]])
climate_only = future_climate - current_cliamte


current_landuse = np.array([[0, 0.3], [0.9, 0]])
future_landuse = np.array([[0, 0.5], [0.7, 0.2]])

current_climate_landuse = current_climate * current_landuse
future_climate_landuse = future_climate * future_landuse
climate_landuse=future_climate_landuse - current_climate_landuse

land_use_only =climate_landuse- climate_only
```

```python
fig, axes = plt.subplots(1, 3, figsize=(9, 2.5))


sns.heatmap(future_climate, ax=axes[0], cmap='Blues', annot=True, cbar=False, fmt='.2f', annot_kws={"size": 12}, square=True)
axes[0].set_title('Future Climate', fontsize=12)

sns.heatmap(current_cliamte, ax=axes[1], cmap='Blues', annot=True, cbar=False, fmt='.2f', annot_kws={"size": 12}, square=True)
axes[1].set_title('Current Climate', fontsize=12)

# Plot SSP4-RCP6.0 heatmaps
sns.heatmap(climate_only, ax=axes[2], cmap='Blues', annot=True, cbar=False, fmt='.2f', annot_kws={"size": 12}, square=True)
axes[2].set_title('Climate Change only', fontsize=12)


# Add equal signs between the plots
fig.text(0.33, 0.5, '-', fontsize=16, ha='center', va='center')
fig.text(0.65, 0.5, '=', fontsize=16, ha='center', va='center')

# Adjust layout and remove ticks
for ax in axes.flat:
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout(rect=[0, 0, 1, 0.92])
plt.show()
```

```python


fig, axes = plt.subplots(1, 2, figsize=(6, 2.5))


sns.heatmap(future_landuse, ax=axes[0], cmap='Greens', annot=True, cbar=False, fmt='.2f', annot_kws={"size": 12}, square=True)
axes[0].set_title('Future Land Use', fontsize=12)

sns.heatmap(current_landuse, ax=axes[1], cmap='Greens', annot=True, cbar=False, fmt='.2f', annot_kws={"size": 12}, square=True)
axes[1].set_title('Current Land Use ', fontsize=12)


# Adjust layout and remove ticks
for ax in axes.flat:
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout(rect=[0, 0, 1, 0.92])
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(9, 2.5))


sns.heatmap(future_climate_landuse, ax=axes[0], cmap='Reds', annot=True, cbar=False, fmt='.2f', annot_kws={"size": 12}, square=True)
axes[0].set_title('Future Climate and Land Use', fontsize=12)

sns.heatmap(current_climate_landuse, ax=axes[1], cmap='Reds', annot=True, cbar=False, fmt='.2f', annot_kws={"size": 12}, square=True)
axes[1].set_title('Current Climate and Land Use', fontsize=12)

# Plot SSP4-RCP6.0 heatmaps
sns.heatmap(climate_landuse, ax=axes[2], cmap='Reds', annot=True, cbar=False, fmt='.2f', annot_kws={"size": 12}, square=True)
axes[2].set_title('Climate and Land Use Change', fontsize=12)


# Add equal signs between the plots
fig.text(0.33, 0.5, '-', fontsize=16, ha='center', va='center')
fig.text(0.65, 0.5, '=', fontsize=16, ha='center', va='center')

# Adjust layout and remove ticks
for ax in axes.flat:
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout(rect=[0, 0, 1, 0.92])
plt.show()
```
